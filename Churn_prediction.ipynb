```python
df1=df1.replace("No phone service","No")
df1=df1.replace("No internet service","No")
```

```python
yes_no_columns=[]
for col in df1:
    df1[col].replace({"Yes":1,"No":0}, inplace=True)

df1
```

```python
df1["gender"].replace({"Female":1,"Male":0},inplace=True)
df1
```

```python
df1=pd.get_dummies(data=df1, columns=["InternetService","Contract","PaymentMethod"])
df1
```

```python
true_false_columns=[]
for col in df1:
    df1[col].replace({"True":1,"False":0}, inplace=True)

df1
```

```python
from sklearn.preprocessing import MinMaxScaler
sk=MinMaxScaler()
Scaled_columns=["tenure","MonthlyCharges","TotalCharges"]

df1[Scaled_columns]=sk.fit_transform(df1[Scaled_columns])
df1
```

```python
x=df1.drop(["Churn","customerID"],axis="columns")
x
y=df1["Churn"]
y
```

```python
df1=df1.replace("No internet service","No")
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model=keras.Sequential([
    keras.layers.Dense(20,input_shape=(26,),activation="relu"),
    keras.layers.Dense(1,activation="sigmoid"),
])
model.compile(optimizer="adam",
              loss="binary_crossentropy",
              metrics=["accuracy"])
model.fit(x_train,y_train,epochs=50)
```

```python
yp=model.predict(x_test)
```

```python
y_test
```

```python
y_p=[]
for a in yp:
    if a>0.5:
        y_p.append(1)
    else:
        y_p.append(0)
    
```

```python
y_p
```

```python
y_test[:5]
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test,y_p))
```

```python
y_p
```

```python
import seaborn as sn
cm=tf.math.confusion_matrix (labels=y_test,predictions=y_p)
cm
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True, fmt='d', linecolor='green', cbar=True, vmin=None, vmax=None,cmap='seismic')
plt.xlabel("Predicted")
plt.ylabel("Truth")
plt.title("Confusion Matrix")
```
